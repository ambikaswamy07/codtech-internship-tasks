```python
import pandas as pd
```

```python
data = {
    "text": [
        "I love this product, it is amazing!",
        "This is the worst experience ever.",
        "The service was okay, not great.",
        "Absolutely fantastic! Highly recommend.",
        "I am very disappointed with the quality.",
        "It works fine but could be better.",
        "Terrible customer support.",
        "Great value for money!"
    ]
}

df = pd.DataFrame(data)
df
```

```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df["Sentiment"] = df["text"].apply(get_sentiment)
df
```

```python
import matplotlib.pyplot as plt
```

```python
sentiment_counts = df["Sentiment"].value_counts()
sentiment_counts
```

```python
sentiment_counts.plot(kind="bar")
plt.title("Sentiment Analysis Results")
plt.xlabel("Sentiment")
plt.ylabel("Number of Texts")
plt.show()
```

## Insights

- The dataset contains a mix of positive, negative, and neutral sentiments.
- Positive sentiments indicate user satisfaction and appreciation.
- Negative sentiments highlight areas where improvement is needed.
- Neutral sentiments represent average or balanced opinions.



## Conclusion

In this task, sentiment analysis was performed on textual data using Natural Language Processing techniques.
The TextBlob library was used to classify text into positive, negative, and neutral categories.
The results were visualized using a bar chart to better understand sentiment distribution.
