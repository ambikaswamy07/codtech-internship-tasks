```python
import pandas as pd
import plotly.express as px

# Load sample dataset
df = px.data.gapminder()

# Filter recent years
df = df[df['year'] >= 2000]

# Create interactive dashboard charts
fig1 = px.line(df, x='year', y='lifeExp', color='continent',
               title='Life Expectancy Over Time')

fig2 = px.bar(df, x='continent', y='pop', color='continent',
              title='Population by Continent', animation_frame='year')

fig1.show()
fig2.show()
```
