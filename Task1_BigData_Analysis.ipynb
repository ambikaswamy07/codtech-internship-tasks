```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("BigDataTask1") \
    .getOrCreate()
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv"
df_pandas = pd.read_csv(url)

df = spark.createDataFrame(df_pandas)
df.show(5)
```

```python
df.printSchema()
```

```python
df.count()
```

```python
df.groupBy("year").avg("passengers").show()
```

```python
df.orderBy(df.passengers.desc()).show(5)
```

INSIGHTS:

1. The dataset shows a clear trend in passenger growth over the years.
2. GroupBy and aggregation operations were efficiently handled using PySpark.
3. Spark enables scalable processing of large datasets compared to traditional methods.
4. This analysis demonstrates how big data tools can extract meaningful insights from large datasets.
